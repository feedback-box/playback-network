```python
AWS_ACCESS_KEY_ID='REDACTED'
AWS_SECRET_ACCESS_KEY='REDACTED'
TABLE_NAME = 'REDACTED'
```

```python
import boto3
from botocore.exceptions import ClientError

# Initialize a session using Amazon DynamoDB
session = boto3.Session(
    aws_access_key_id=AWS_ACCESS_KEY_ID,
    aws_secret_access_key=AWS_SECRET_ACCESS_KEY,
    region_name='ap-southeast-2'
)

# Initialize DynamoDB resource
dynamodb = session.resource('dynamodb')

# Initialize DynamoDB client
client = session.client('dynamodb')

# List all tables to verify the connection
try:
    response = client.list_tables()
    print("Tables in DynamoDB:")
    for table in response['TableNames']:
        print(table)
except ClientError as e:
    print(f"Error listing tables: {e}")

# Select your DynamoDB table
table_name = TABLE_NAME

# Debug: Print table name and region
print(f"Table Name: {table_name}")
print(f"Region: ap-southeast-2")

table = dynamodb.Table(table_name)

# Function to scan the table and retrieve all items
def get_all_items(table):
    try:
        response = table.scan()
        items = response['Items']
        
        # Handle pagination in case of large number of items
        while 'LastEvaluatedKey' in response:
            response = table.scan(ExclusiveStartKey=response['LastEvaluatedKey'])
            items.extend(response['Items'])
        
        return items
    except ClientError as e:
        print(f"Error: {e}")
        raise

# Retrieve all items from the table
try:
    items = get_all_items(table)
    # Print retrieved items
    for item in items:
        print(item)
except ClientError as e:
    print(f"Failed to retrieve items: {e}")
```

```python
import boto3
from botocore.exceptions import ClientError
import csv
import uuid
from datetime import datetime

# Initialize a session using Amazon DynamoDB
session = boto3.Session(
    aws_access_key_id=AWS_ACCESS_KEY_ID,
    aws_secret_access_key=AWS_SECRET_ACCESS_KEY,
    region_name='ap-southeast-2'
)

# Initialize DynamoDB resource and client
dynamodb = session.resource('dynamodb')
client = session.client('dynamodb')

# List all tables to verify the connection
try:
    response = client.list_tables()
    print("Tables in DynamoDB:")
    for table in response['TableNames']:
        print(table)
except ClientError as e:
    print(f"Error listing tables: {e}")
    exit(1)

# Select your DynamoDB table
table_name = TABLE_NAME

# Debug: Print table name and region
print(f"Table Name: {table_name}")
print(f"Region: ap-southeast-2")

table = dynamodb.Table(table_name)

# Read the CSV file and get the first row
csv_file_path = 'TASKS_20240602_CLEAN.csv'

with open(csv_file_path, mode='r') as file:
    csv_reader = csv.DictReader(file)
    first_row = next(csv_reader)

# Prepare the item to be inserted
current_time = datetime.utcnow().isoformat()
item = {
    'id': str(uuid.uuid4()),  # Generate a unique ID
    'app': first_row['App'].upper(),
    'description': first_row['Task'].upper(),  # Set description equal to the task
    'difficulty': 1,  # Set difficulty to 1 for all tasks
    'name': first_row['Task'].upper(),  # Set name to be the same as the task
    'createdAt': current_time,  # Current timestamp in ISO format
    'updatedAt': current_time,   # Current timestamp in ISO format
    '__typename': 'Task',
    'walletAddress': '',
    'appImage': first_row['Index'] + '.jpg'
}

# Function to insert an item into the table
def insert_item(table, item):
    try:
        table.put_item(Item=item)
        print(f"Successfully inserted item: {item}")
    except ClientError as e:
        print(f"Error inserting item: {e}")

# Insert the first row into the table
insert_item(table, item)
```
