# Load data from folder and combine into single CSV

```python
import pandas as pd
import os

# Set the path to your folder
folder_path = "TASKS_CLEANED"

# Create an empty list to store DataFrames
dfs = []

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    # Only process CSV files
    if filename.endswith(".csv"):
        # Construct the full file path
        file_path = os.path.join(folder_path, filename)
        print(file_path)
        
        # Read the CSV into a DataFrame
        df = pd.read_csv(file_path, usecols=[0, 1, 2, 3, 4])
        df.columns = "Index,Category,Subcategory,App,Task".split(',')
        
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.drop_duplicates(inplace=True)
combined_df.to_csv("cleaned.csv", index=False)

print("All CSV files combined into 'cleaned.csv'")
```
